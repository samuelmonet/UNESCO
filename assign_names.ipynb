```python
for i in range(441):
    if kobo.loc[i]['label::English(en)']==kobo.loc[i]['label::English(en)']:
        for quest in questions.index:
            if kobo.loc[i]['label::English(en)']==quest:
                questions.at[quest,'Idquest']=kobo.loc[i]['name']
                if kobo.loc[i]['type']=='decimal':
                    questions.at[quest,'type']='continue'
            elif kobo.loc[i]['label::English(en)'] in quest and '/' in quest:
                questions.at[quest,'Idquest']=kobo.loc[i]['name']+' '+quest.split('/')[-1]
                questions.at[quest,'Treatment']='cat'
            
```

```python
questions.T.to_csv('questions_auto.csv',sep='\t')
```
